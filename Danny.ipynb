#  Line plot


```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plot Basics

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,3,5,7])

plt.plot(
    x, y,
    color = "red",
    linewidth = 2,
    linestyle = "-",
    marker = "o",
    markersize = 8,
    label = "Plot legend"
)

plt.grid(True)

plt.xlabel("X Axis Label")
plt.ylabel("Y Axis Label")

plt.suptitle("Main Plot Title", fontsize=16)
plt.title("Subtitle describing the chart", fontsize=10)

plt.legend()


plt.xlim(0, 6)
plt.ylim(0, 8)
# Annotation of point
plt.annotate(
    "Data point",
    xy = (3,3),
    xytext = (3,6),
    arrowprops = dict(arrowstyle="->")
)

plt.annotate(
    "Y Axis",
    xy = (0,6),
    xytext = (1,6),
    arrowprops = dict(arrowstyle="->")
)

plt.annotate(
    "X Axis",
    xy = (4,0),
    xytext = (4,1),
    arrowprops = dict(arrowstyle="->")
)

plt.annotate(
    "Grid",
    xy = (2,2),
    xytext = (2.5,2.2),
    arrowprops = dict(arrowstyle="->")
)


plt.show()
```

## Adding a little more spice to the plot

```python
# ---------------------------
# DATA
# ---------------------------

x = np.array([1,2,3,4,5])          # X-axis values
y = np.array([2,4,3,5,7])          # Y-axis values

# ---------------------------
# FIGURE SETUP
# ---------------------------

plt.figure(
    num = "Example Figure",          # Optional name for the figure window
    figsize = (10,6),                # Width, Height in inches
    dpi = 100,                       # Resolution of the figure
    facecolor = "white",             # Background color of the figure
    edgecolor = "black"              # Border color of the figure
)

# ---------------------------
# MAIN PLOT
# ---------------------------

plt.plot(
    x, y,

    # Line appearance
    color = "blue",                  # Line color
    linestyle = "-",                 # Line style (-, --, -., :)
    linewidth = 2,                   # Thickness of the line

    # Marker settings
    marker = "o",                    # Marker style (o, s, ^, *, etc.)
    markersize = 8,                  # Marker size
    markerfacecolor = "red",         # Marker fill color
    markeredgecolor = "black",       # Marker outline color
    markeredgewidth = 1,             # Marker outline width

    # Transparency
    alpha = 0.9,                     # Transparency (0 = invisible, 1 = solid)

    # Label for legend
    label = "Sample Data"            # Label used in legend
)

# ---------------------------
# TITLES
# ---------------------------

plt.suptitle(
    "Main Plot Title",             # Title text
    fontsize = 18,                   # Title font size
    fontweight = "bold",             # Font weight
    color = "black",                 # Title color
)

# NOTE: Matplotlib has no official subtitle argument,
# but you can simulate it using fig.text()

plt.title(
    "Subtitle describing the chart",
    ha = "center",                   # Horizontal alignment
    fontsize = 12
)

# ---------------------------
# AXIS LABELS
# ---------------------------

plt.xlabel(
    "X Axis Label",                # Sets the label for the x axis
    fontsize = 12,
    labelpad = 10                    # Spacing between axis and label
)

plt.ylabel(
    "Y Axis Label",                # Sets the label for the y axis
    fontsize = 12,
    labelpad = 10
)

# ---------------------------
# AXIS LIMITS
# ---------------------------

plt.xlim(0, 6)                    # Minimum and maximum values for X axis
plt.ylim(0, 10)                   # Minimum and maximum values for Y axis

# ---------------------------
# TICKS
# ---------------------------

plt.xticks(
    ticks = [1,2,3,4,5],             # Locations of tick marks
    labels = ["A","B","C","D","E"],  # Custom labels for ticks
    rotation = 45,                   # Rotate tick labels
    fontsize = 10
)

plt.yticks(
    fontsize=10                    # Font size of Y ticks
)

# ---------------------------
# GRID
# ---------------------------

plt.grid(
    True,                          # Turns grid on/off
    which = "both",                  # Major, minor, or both
    axis = "both",                   # X, Y, or both
    linestyle = "--",                # Grid line style
    linewidth = 0.5,
    alpha = 0.7
)

# ---------------------------
# LEGEND
# ---------------------------

plt.legend(
    loc = "upper left",              # Legend location
    fontsize = 10,                   # Font size
    frameon = True,                  # Draw box around legend
    shadow = True,                   # Add shadow
    title = "Legend Title"           # Optional legend title
)

# ---------------------------
# ANNOTATIONS
# ---------------------------

plt.annotate(
    "Important Point",             # Text to display
    xy = (3,3.1),                  # Point being annotated
    xytext = (3.2,7),              # Location of annotation text
    arrowprops = dict(
        arrowstyle = "->",         # Arrow style
        color = "black"
    ),
    fontsize = 10
)

# ---------------------------
# REFERENCE LINES
# ---------------------------

plt.axhline(
    y=5,                           # Horizontal reference line value
    color = "red",
    linestyle = "--",
    linewidth = 1
)

plt.axvline(
    x = 3,                           # Vertical reference line value
    color = "green",
    linestyle = ":",
    linewidth = 1
)

# ---------------------------
# FILL AREA
# ---------------------------

plt.fill_between(
    x, y,
    color = "lightblue",             # Fill color
    alpha = 0.3                      # Fill transparency
)

# ---------------------------
# LAYOUT ADJUSTMENTS
# ---------------------------

plt.tight_layout()                 # Automatically adjust spacing
plt.subplots_adjust(
    left = 0.1,                      # Left margin
    right = 0.9,                     # Right margin
    top = 0.9,                       # Top margin
    bottom = 0.1                     # Bottom margin
)

# ---------------------------
# SAVE OUTPUT
# ---------------------------

plt.savefig(
    "plot.png",                    # File name
    dpi = 300,                       # Export resolution
    bbox_inches = "tight",           # Trim extra whitespace
    facecolor = "white"
)

# ---------------------------
# DISPLAY
# ---------------------------

plt.show()                         # Renders the plot

```
